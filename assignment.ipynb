```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

import seaborn as sns
```

```python
train_data=pd.read_csv("dataset/training_set.csv")
test_data=pd.read_csv("dataset/test_set.csv")
```

# EDA AND PREPROCESSING

```python
train_data.info()
```

```python
train_data.head()
```

```python
train_data=pd.read_csv("dataset/training_set.csv",index_col=0)
test_data=pd.read_csv("dataset/test_set.csv",index_col=0)
```

```python
train_data.head()
```

```python
train_data.isnull().sum
```

```python

for i in range(1,57):
    col="X"+str(i)
    train_data[col]=train_data[col].replace(0,np.NAN)
    train_data[col].fillna(train_data[col].mean(), inplace=True)
res=train_data.isnull().sum()
print(res)
```

```python
train_data.head()
```

```python
test_data=test_data.replace(0,np.NAN)
for i in range(1,57):
    col="X"+str(i)
    test_data[col].fillna(test_data[col].mean(), inplace=True)
```

```python
test_data.head()
```

# FEATURE SELECTION

```python
features=train_data.drop('Y',axis="columns")
target=train_data['Y']
train_data.head()
```

```python
features.head()
```

```python
target.head()
```

```python
#splitting training data into 4:1 training data and validation data
x_train, x_test, y_train, y_test = train_test_split(
  features,target ,test_size=0.25)
```

```python
#FEATURE SCALING
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
model=KNeighborsClassifier(n_neighbors=3)
model.fit(x_train,y_train)
```

```python
#prediction on validation data
y_pred=model.predict(x_test)
```

```python
#finding accuracy
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
